```python
from gurobipy import *
import numpy as np
import pandas as pd

#########Parameters Set-up############
# Read the data from the csv file and use the first column as the index of rows
data = pd.read_csv('OR Project Data.csv')#specify the first column as index otherwise treated as data
print(data.shape)

# Record the number of rows and columns in the data
temp_M, temp_N = data.shape
```

```python
#Constraint right-hand-side: capacity for each plant
nutrient = data.iloc[0:temp_M, 2:temp_N]
nutrient = nutrient.values
```

```python
req = pd.read_csv('OR Project Data Constraints.csv')#specify the first column as index otherwise treated as data
req
```

```python
men_req = req.iloc[0:8, 2]
women_req = req.iloc[0:8, 3]
```

```python
req
```

```python
#########Model Set-up###############

m = Model("diet")

# Creat variables
# addVars ( *indices, lb=0.0, ub=GRB.INFINITY, obj=0.0, vtype=GRB.CONTINUOUS, name="" )
x = m.addVars(temp_M,vtype=GRB.BINARY, name = 'meal')

# Set objective
m.setObjective((quicksum(x[i] for i in range(121))), GRB.MINIMIZE)

# Add constraints: 
m.addConstrs(quicksum(x[i] for i in range(59)) == 2 , name = "main")
m.addConstrs(quicksum(x[i] for i in range(94,121)) == 1 , name = "breakfast")
m.addConstrs(quicksum(x[i] for i in range(59,94)) <= 2 , name = "optional")

m.addConstrs(( quicksum(nutrient[i,j]*x[i] for i in range(121)) >= men_req[j] for j in range(8) ), name = "requirement")

# Solving the model
m.optimize()

#  Print optimal solutions and optimal value
for v in m.getVars():
    print(v.VarName, v.x)
    
print('Obj:', m.objVal)
```

# Code for Drug Production Problem

### LP Formulation for the nominal problem:
\begin{equation}
\begin{split}
\max~ & (6200-700)x - 100y_1 - 199.9y_2 \\
s.t. ~& 90x\leq 2000, \quad\text{(manpower)}\\
& 40x\leq 800, \quad \text{(equippment)}\\
& y_1+y_2\leq 1000, \quad \text{(storage)}\\
& 700x + 100y_1 + 199.9y_2\leq 100000, \quad \text{(budget)}\\
& 0.01y_1 + 0.02y_2 - 0.5x\geq 0, \quad \text{(ingredient)}\\
& x,y_1,y_2\geq 0.
\end{split}
\end{equation}

```python
from gurobipy import *
import numpy as np

#########Parameters Set-up############

#price for each product
price = np.array([6200])
#cost for each product
cost_prod = np.array([700])
#the amount of ingredient required by each product
ingre_prod = np.array([0.5])
#number of products
N_PROD = len(cost_prod)

#the resources available for production, e.g., manpower and equippment
res_avail = np.array([2000, 800])
#the amount of resource i consumed by producing one unite of product j 
res_coef = np.array([[90],
                    [40]])
#number of resources
N_RES = len(res_avail)

#cost for each source of raw material
cost_raw = np.array([100, 199.9])
#the amount of ingredient contained in each source of raw material
ingre_raw = np.array([0.01, 0.02])
#number of sources for raw material
N_RAW = len(cost_raw)

#storage capacity
storage = 1000

#total budget
budget = 100000

print('Number of Products %g; Number of resources %g; Number of raw materials %g' % (N_PROD, N_RES, N_RAW))
```

```python
#########Model Set-up###############

m = Model("drug")

# production decisions
x_prod = m.addVars(N_PROD, name = "x_prod")

# procurment decisions
x_raw = m.addVars(N_RAW, name = "x_raw")

# maximize the revenue minus procurement cost
m.setObjective( quicksum((price[i] - cost_prod[i])*x_prod[i] for i in range(N_PROD)) - 
               quicksum(cost_raw[j]*x_raw[j] for j in range(N_RAW)), GRB.MAXIMIZE)

# resource constraints
m.addConstrs(( quicksum(res_coef[i,j]*x_prod[j] for j in range(N_PROD)) <= res_avail[i] for i in range(N_RES) ), "Resource")

# ingredient constraint
m.addConstr( quicksum(ingre_raw[j]*x_raw[j] for j in range(N_RAW)) 
            - quicksum(ingre_prod[i]*x_prod[i] for i in range(N_PROD)) >= 0, "Ingredient")

# storage constraint
m.addConstr( quicksum(x_raw[j] for j in range(N_RAW)) <= storage, "Storage")

# budget constraint
m.addConstr( quicksum(cost_prod[i]*x_prod[i] for i in range(N_PROD)) 
            + quicksum(cost_raw[j]*x_raw[j] for j in range(N_RAW)) <= budget, "Budget")

# Solving the model
m.optimize()

#  Print optimal solutions and optimal value
print("\n Production and procurement decisions:")
for v in m.getVars():
    print(v.VarName, v.x)

print("\n Optimal profit:")
print('Obj:', m.objVal)

#  Print optimal dual solutions
print("\n Dual solutions:")
for d in m.getConstrs():
    print('%s %g' % (d.ConstrName, d.Pi))
```

### LP Formulation for the worst-case problem:
\begin{equation}
\begin{split}
\max~ & (6200-700)x - 100y_1 - 199.9y_2 \\
s.t. ~& 90x\leq 2000, \quad\text{(manpower)}\\
& 40x\leq 800, \quad \text{(equippment)}\\
& y_1+y_2\leq 1000, \quad \text{(storage)}\\
& 700x + 100y_1 + 199.9y_2\leq 100000, \quad \text{(budget)}\\
& 0.00995y_1 + 0.0196y_2 - 0.5x\geq 0, \quad \text{(ingredient)}\\
& x,y_1,y_2\geq 0.
\end{split}
\end{equation}

```python
#the amount of ingredient contained in each source of raw material in the worst case
ingre_raw = np.array([0.00995, 0.0196])

#extract the ingredient constraint that is to be modified
C_ingredient = m.getConstrByName("Ingredient")

#modify the coefficient of x_raw 
for i in range(N_RAW):
    m.chgCoeff(C_ingredient, x_raw[i], ingre_raw[i])

# Solving the model
m.optimize()

#  Print optimal solutions and optimal value
print("\n Production and procurement decisions:")
for v in m.getVars():
    print(v.VarName, v.x)

print("\n Optimal profit:")
print('Obj:', m.objVal)

```

### LP Formulation for the worst-case problem with conservativeness control:
\begin{equation}
\begin{split}
\max~ & (6200-700)x - 100y_1 - 199.9y_2 \\
s.t. ~& 90x\leq 2000, \quad\text{(manpower)}\\
& 40x\leq 800, \quad \text{(equippment)}\\
& y_1+y_2\leq 1000, \quad \text{(storage)}\\
& 700x + 100y_1 + 199.9y_2\leq 100000, \quad \text{(budget)}\\
& 
\left.
\begin{aligned}
& 1\%y_1 + 2\%y_2 + 2\alpha p_1 + p_6 + p_7 - 0.5x\geq 0,\\
&p_2 - p_3 = 0.005\%y_1,\\
&p_4 - p_5 = 0.04\%y_2,\\
&p_1 - p_2 - p_3 + p_6 \leq 0,\\
&p_1 - p_4 - p_5 + p_7 \leq 0,\\
\end{aligned}
\right\} \quad \text{(ingredient)} \\
& x,y_1,y_2\geq 0, p_i\leq 0, i=1,...,7.
\end{split}
\end{equation}

```python
#########Model Set-up (Lazy Version without Separating Data and Model)###############

#parameter for controlling conservativeness level
alpha = 1

m_robust = Model("drug_rosbut")

# production decisions
x_prod = m_robust.addVar(name = "x_prod")

# procurment decisions
x_raw = m_robust.addVars(2, name = "x_raw")

# dual variables
p_dual = m_robust.addVars(7, lb=-GRB.INFINITY, ub=0.0, name = "p_dual")

# Set objective
m_robust.setObjective((6200-700)*x_prod - 100*x_raw[0] - 199.9*x_raw[1], GRB.MAXIMIZE)

# Add constraint: 
m_robust.addConstr(90*x_prod <= 2000, "Manpower")

m_robust.addConstr(40*x_prod <= 800, "Equippment")

m_robust.addConstr(x_raw[0] + x_raw[1] <= 1000, "Storage")

m_robust.addConstr(100*x_raw[0] + 199.9*x_raw[1] + 700*x_prod <= 100000, "Budget")

m_robust.addConstr(0.01*x_raw[0] + 0.02*x_raw[1] + 2*alpha*p_dual[0]  + 
                   p_dual[5] + p_dual[6] - 0.5*x_prod >= 0, "Ingredient")

m_robust.addConstr(p_dual[1] - p_dual[2] == 0.00005*x_raw[0], "Dual_Cons_1")

m_robust.addConstr(p_dual[3] - p_dual[4] == 0.0004*x_raw[1], "Dual_Cons_2")

m_robust.addConstr(p_dual[0] - p_dual[1] - p_dual[2] + p_dual[5] <= 0, "Dual_Cons_3")

m_robust.addConstr(p_dual[0] - p_dual[3] - p_dual[4] + p_dual[6] <= 0, "Dual_Cons_4")

# Solving the model
m_robust.optimize()

#  Print optimal solutions and optimal value
print("\n Production and procurement decisions:")
for v in m_robust.getVars():
    print(v.VarName, v.x)

print("\n Optimal profit:")
print('Obj:', m_robust.objVal)
```

```python
#########Sensitivity Analysis regarding to alpha###############

#enumerating alpha from 0 to 1 with stepsize 0.2
#if we are very confident, the alpha is closer to 0. 
alpha_v = np.linspace(0.0, 1.0, num=6)

N_alpha = len(alpha_v)

#intitalizing the outputs of profits, production decisions and procurement decisions
profit_v = np.zeros(N_alpha)
x_prod_v = np.zeros(N_alpha)
x_raw_v = np.zeros((2, N_alpha))

#Supressing the optimization output
m_robust.setParam( 'OutputFlag', False )


for i in range(N_alpha):
    
    alpha = alpha_v[i]
    
    #extract the ingredient constraint that is to be modified
    C_ingredient = m_robust.getConstrByName("Ingredient")

    #modify the coefficient of p_dual[0] 
    m_robust.chgCoeff(C_ingredient, p_dual[0], 2*alpha)   

    # Solving the model
    m_robust.optimize()
    
    profit_v[i] = m_robust.objVal
    x_prod_v[i] = x_prod.x
    x_raw_v[0,i] = x_raw[0].x
    x_raw_v[1,i] = x_raw[1].x



```

```python
#########Visualizing the Results###############

import matplotlib.pyplot as plt

#Plot the worst case profit along with production decisions
plt.subplot(121)
plt.plot(alpha_v, profit_v)
plt.xlabel(r'$\alpha$')
plt.ylabel('worst case profit')

plt.subplot(122)
plt.plot(alpha_v, x_prod_v)
plt.xlabel(r'$\alpha$')
plt.ylabel('production quantities')

plt.tight_layout()
plt.show()    

#Plot the procurement decisions
plt.plot(alpha_v, x_raw_v[0,:], 'r--', alpha_v, x_raw_v[1,:], 'b-.')
plt.xlabel(r'$\alpha$')
plt.ylabel('procurement quantities')
plt.legend(('Source 1', 'Source 2'))
plt.show() 
   
```
